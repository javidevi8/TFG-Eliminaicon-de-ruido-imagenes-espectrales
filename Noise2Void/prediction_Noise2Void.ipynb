# Comprobacion de resultados

```python
# importamos dependencias
from n2v.models import N2V
import numpy as np
from matplotlib import pyplot as plt
from tifffile import imread
from csbdeep.io import save_tiff_imagej_compatible
from matplotlib import image
```

## Load the Network

```python
# Cargamos el modelo que queremos, creando un uevo modelo sin crear una configuración.
model_name = 'n2v_ALBA_Prueba_2_Tensor'
basedir = 'models'
model = N2V(config=None, name=model_name, basedir=basedir)
```

```python
# Se pueden cargar los mejores pesos o los pesos finales del modelo con el método load.weights
# model.load_weights('weights_last.h5')
```

## Prediction


```python
#Configuramos el modelo con las mismas dimensiones que  con las que hemnos entrenado
model.config.axes='YXC'
```

```python
# Cargamos las imagenes que queremos comprobar con solo 3 dimensiones
img = image.imread('Datos_entreno/dimension0_143.png')[...,:3]
img.astype(np.float32)


pred = model.predict(img, axes='YXC')
```

```python
# Vemos los resultados
plt.figure(figsize=(30,30))

plt.subplot(1,2,1)
plt.imshow( img)
plt.title('Input')

plt.subplot(1,2,2)
plt.imshow( pred )
plt.title('Prediction')

plt.show()
```
